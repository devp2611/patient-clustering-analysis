CONVERT TO DATAFRAME FOR SUBJECT S2

```python
import pickle
import pandas as pd

# Load the .pkl file using the 'latin1' encoding
with open("/content/drive/MyDrive/FDA PROJECT 2/S2/S2.pkl", "rb") as file:
    data = pickle.load(file, encoding='latin1')  # Specify 'latin1' encoding

#print(data)  # Display the contents
S2 = data # Assign the loaded data to S2
data4 = S2['signal']['chest']['ACC']
dfS2 = pd.DataFrame(data = data4, columns = ['acc_x', 'acc_y', 'acc_z'])

def add_multiple_columns(dfS2, column_names, S2, S2_column):
    for col in column_names:
        dfS2[col] = S2['signal']['chest'][col]
    return dfS2

column_names = ['ECG', 'EMG', 'EDA', 'Temp', 'Resp']
dfS2 = add_multiple_columns(dfS2, column_names, S2, 'ECG')
dfS2['State'] = S2.get('label')
dfS2['Subject'] = S2.get('subject')
dfS2
```
